```python
#LIBRARY
import json
import pandas as pd
import string
import nltk
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
#JSON
with open('D:/data/ww1_letters/letters.json', 'r', encoding="utf8") as f:
    data = json.load(f)
    letter1 = data["na_uk_02"]
```

```python
#REMOVE PUNCTUATION
textclean_0 = "".join([i.lower() for i in letter1 if i not in string.punctuation])
textclean_0
```

```python
#TOKENIZATION
stopwordss = nltk.corpus.stopwords.words('english')
textclean_1 = ''.join([i for i in textclean_0 if i not in string.punctuation])
textclean_2 = nltk.tokenize.word_tokenize(textclean_1)
textclean_2
```

```python
textclean_3 = [i for i in textclean_2 if i not in stopwordss]
print(textclean_3)
```

```python
#STEMMING
ps = nltk.PorterStemmer()
textclean_4 = [ps.stem(word) for word in textclean_3]
```

```python
#LEMMATIZE
wn = nltk.WordNetLemmatizer()
textclean_5 = [wn.lemmatize(word) for word in textclean_4]
textclean_5
```

```python
#JSON BYTE TO STRING
def listToString(s):
    str1 = " "
    return (str1.join(s))

textclean_6 = listToString(textclean_5)
type(textclean_6)
```

```python
#WORDCLOUD
letter1_wc = WordCloud(
    background_color='white',
    max_words=20
)

letter1_wc.generate(textclean_6)
```

```python
plt.imshow(letter1_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
